```python
import pandas as pd
import numpy as np
import random
import os
```

```python
masterDF = pd.DataFrame()

for month in ["01", "02","03","04","05","06","07","08","09","10","11","12"]:
    filepath = f"Data/2019{month}-citibike-tripdata.csv"
    
    df_input = pd.read_csv(filepath, low_memory=False)
    sample = df_input.sample(frac=0.02, random_state=42)
    masterDF = masterDF.append(sample)
    
    print(month)
```

```python
masterDF.reset_index(drop=True, inplace=True)

masterDF.to_csv("full2019sample.csv", index=False)
```
